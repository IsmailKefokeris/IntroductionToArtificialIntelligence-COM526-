```python
import pandas as pd
import numpy as np
import plotly.express as px
```

```python
world_data = pd.read_csv("countries of the world.csv")
world_data.head()
```

```python
fig = px.treemap(world_data,
                path = ["Region", "Country", "Population", "Deathrate"],
                values = "Population",
                color = "GDP ($ per capita)",
                color_continuous_scale = "RdBu",
                width=1000, height=600,
                title="World Population Distribution Color-encoded by GDP",)

fig.show()
```

```python
df = pd.read_csv("vgsales.csv")
df.dropna(inplace=True)
df
```

```python
game_fig = px.treemap(df,
                path = ["Platform", "Genre"],
                values = "Global_Sales",
                color = "NA_Sales",
                color_continuous_scale = "RdBu",
                width=1000, height=700,
                title="Tree map of Global Game Sales",)

game_fig.show()
```
